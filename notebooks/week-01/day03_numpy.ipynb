```python
x = np.arange(1, 11)
train_loss = [0.9, 0.8, 0.7, 0.6, 0.5, 0.45, 0.4, 0.38, 0.35, 0.33]
val_loss   = [0.95, 0.85, 0.78, 0.72, 0.68, 0.66, 0.65, 0.64, 0.64, 0.65]

plt.plot(x, train_loss, label='Train Loss', color='blue')
plt.plot(x, val_loss,   label='Val Loss',   color='red')
plt.title("Training vs Validation Loss")
plt.xlabel("Epoch")
plt.ylabel("Loss")
plt.legend()
plt.grid(True)
plt.show()
```

```python
epochs = np.arange(1, 21)
train_loss = 1 / (1 + np.log(epochs)) + np.random.rand(20) * 0.02
val_loss   = 1 / (1 + np.log(epochs)) + np.random.rand(20) * 0.05 + 0.05

best_epoch = np.argmin(val_loss)

plt.figure(figsize=(10, 5))
plt.plot(epochs, train_loss, label='Train Loss', color='blue', linewidth=2)
plt.plot(epochs, val_loss,   label='Val Loss',   color='red',  linewidth=2)
plt.axvline(x=best_epoch+1, color='green', linestyle='--', label='Best Epoch')
plt.scatter(best_epoch+1, val_loss[best_epoch], color='green', s=100, zorder=5)
plt.title("Training vs Validation Loss")
plt.xlabel("Epoch")
plt.ylabel("Loss")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
raw_data    = np.random.exponential(scale=2, size=1000)
scaled_data = (raw_data - raw_data.mean()) / raw_data.std()

fig, axes = plt.subplots(1, 2, figsize=(12, 4))

axes[0].hist(raw_data,    bins=40, color='tomato',    edgecolor='black', alpha=0.7)
axes[0].set_title("Before Scaling")
axes[0].set_xlabel("Value")
axes[0].set_ylabel("Frequency")

axes[1].hist(scaled_data, bins=40, color='steelblue', edgecolor='black', alpha=0.7)
axes[1].set_title("After Scaling")
axes[1].set_xlabel("Value")

plt.suptitle("Feature Distribution Before vs After Scaling", fontsize=14)
plt.tight_layout()
plt.show()
```

```python
time     = np.arange(100)
cpu      = 30 + np.random.randn(100) * 5
# inject anomalies
cpu[20]  = 90
cpu[55]  = 95
cpu[80]  = 88

threshold   = cpu.mean() + 2 * cpu.std()
anomalies   = time[cpu > threshold]
anom_values = cpu[cpu > threshold]

plt.figure(figsize=(12, 5))
plt.plot(time, cpu, color='steelblue', linewidth=1.5, label='CPU Usage')
plt.axhline(y=threshold, color='orange', linestyle='--', label=f'Threshold ({threshold:.1f}%)')
plt.scatter(anomalies, anom_values, color='red', s=100, zorder=5, label='Anomaly')
plt.fill_between(time, threshold, cpu, where=(cpu > threshold), color='red', alpha=0.2)
plt.title("CPU Usage — Anomaly Detection")
plt.xlabel("Time")
plt.ylabel("CPU %")
plt.legend()
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()
```

```python
data = np.random.randn(5, 5)
corr = np.corrcoef(data)

fig, ax = plt.subplots(figsize=(7, 6))
im = ax.imshow(corr, cmap='coolwarm', vmin=-1, vmax=1)
plt.colorbar(im)

features = ['CPU', 'Memory', 'Disk', 'Network', 'Latency']
ax.set_xticks(np.arange(5))
ax.set_yticks(np.arange(5))
ax.set_xticklabels(features)
ax.set_yticklabels(features)

# Add values inside boxes
for i in range(5):
    for j in range(5):
        ax.text(j, i, f'{corr[i,j]:.2f}', ha='center', va='center', fontsize=9)

plt.title("Feature Correlation Matrix")
plt.tight_layout()
plt.show()
```

```python
fig, axes = plt.subplots(2, 2, figsize=(14, 10))
fig.suptitle("MLOps Monitoring Dashboard", fontsize=16, fontweight='bold')

# Chart 1 — Training loss
epochs = np.arange(1, 21)
axes[0,0].plot(epochs, 1/np.log(epochs+1), color='blue', linewidth=2)
axes[0,0].set_title("Training Loss")
axes[0,0].set_xlabel("Epoch")
axes[0,0].grid(True, alpha=0.3)

# Chart 2 — CPU anomaly
time = np.arange(100)
cpu  = 30 + np.random.randn(100) * 5
cpu[40] = 92
axes[0,1].plot(time, cpu, color='steelblue')
axes[0,1].axhline(y=cpu.mean() + 2*cpu.std(), color='red', linestyle='--')
axes[0,1].set_title("CPU Anomaly Detection")
axes[0,1].grid(True, alpha=0.3)

# Chart 3 — Data distribution
axes[1,0].hist(np.random.randn(500), bins=30, color='green', alpha=0.7, edgecolor='black')
axes[1,0].set_title("Feature Distribution")
axes[1,0].grid(True, alpha=0.3)

# Chart 4 — Scatter classes
axes[1,1].scatter(np.random.randn(50),     np.random.randn(50),     color='blue', label='Class 0', alpha=0.6)
axes[1,1].scatter(np.random.randn(50)+3,   np.random.randn(50)+3,   color='red',  label='Class 1', alpha=0.6)
axes[1,1].set_title("Class Separation")
axes[1,1].legend()
axes[1,1].grid(True, alpha=0.3)

plt.tight_layout()
plt.show()
```
