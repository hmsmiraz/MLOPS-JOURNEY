```python
import numpy as np

# Normal Python list
py_list = [1, 2, 3, 4, 5]

# NumPy array
arr = np.array([1, 2, 3, 4, 5])

print("List:", py_list)
print("Array:", arr)
print("Shape:", arr.shape)   # how many elements
print("Type:", arr.dtype)    # what kind of numbers
```

```python
# 2 rows, 3 columns
arr2d = np.array([[1, 2, 3],
                  [4, 5, 6]])

print(arr2d)
print("Shape:", arr2d.shape)   # (2, 3) → 2 rows, 3 cols
print("Rows:", arr2d.shape[0])
print("Cols:", arr2d.shape[1])
```

```python
import time

big = np.arange(1000000)  # 1 million numbers

# Slow way — Python loop
start = time.time()
result = [x * 2 for x in big]
print("Loop time:", round(time.time() - start, 4), "sec")

# Fast way — NumPy
start = time.time()
result = big * 2
print("NumPy time:", round(time.time() - start, 4), "sec")
```

```python
import time

sizes = [1000, 10000, 100000, 1000000]

for size in sizes:
    arr = np.arange(size)
    
    # Python loop
    start = time.time()
    result = [x ** 2 for x in arr]
    loop_time = round(time.time() - start, 5)
    
    # NumPy
    start = time.time()
    result = arr ** 2
    numpy_time = round(time.time() - start, 5)
    
    print(f"Size {size:>8} → Loop: {loop_time}s | NumPy: {numpy_time}s")
```
